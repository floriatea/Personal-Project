```python
import itertools
import pandas as pd
import numpy as np
```

```python
import xlrd
df=pd.read_excel("Schema Python dummy data.xlsx")
df.columns = df.iloc[1]
df = df.iloc[2: , :]
df.reset_index(drop=True, inplace=True)
df['datePublished'] = pd.to_datetime(df['datePublished']).dt.date
df['dateModified'] = pd.to_datetime(df['dateModified']).dt.date

df.info()
df.head()
```

```python
#Method 3
#easiest and fastest without installing anything
#if need the filename to be current date, need to import a time package
#drawback : need to change variable order after the '%' accordingly if the order of the code is changed 

import time
import os

for i in range(len(df)):
    
    ids = df['id'][i]
    headline=df['headline'][i]
    description=df['description'][i]
    image=df['image'][i]
    name=df['name'][i]
    url=df['url'][i]
    datePublished=df['datePublished'][i]
    dateModified=df['dateModified'][i]
    
    text= """
    <script type="application/ld+json">

    {
      "@context": "https://schema.org",
      "@type": "Article",
      "mainEntityOfPage": {
       "@type": "WebPage",
       "@id": "%s"
    },
      "headline": "%s",
      "description": "%s",
      "image": "%s",  
      "author": {
       "@type": "Organization",
       "name": "%s"
    },  
      "publisher": {
        "@type": "Organization",
        "name": "%s",
        "logo": {
         "@type": "ImageObject",
         "url": "%s"
     }
    },
      "datePublished": "%s",
      "dateModified": "%s"
    }
    </script>  
    """ % (ids,headline,description,image,name,name,url,datePublished,dateModified)
    print(text)
    
    TodaysDate = time.strftime("%Y-%m-%d")
    #textfilepath = 'C:\Users\flori\Documents'
    textfilename = TodaysDate +"-Article Schema "+ "-"+ df['headline'][i]
    characters_to_remove = "/\:*?<>|"
    pattern = "[" + characters_to_remove + "]"
    textfilename=re.sub(pattern, " ",textfilename)
    filepath = os.path.join(os.getcwd(), textfilename)
    
    f = open(filepath +".txt", "w")
    f.write(text)
    f.close()
```
